## Dropping 'unnamed' Column

```python
df = df.drop('Unnamed: 0', axis = 1)
df
```

## Change weekly study hours column

```python
df['WklyStudyHours'] = df['WklyStudyHours'].str.replace('05-Oct', '5-10')
df
```

## Gender Distribution

```python
plt.figure(figsize=(6,5))
ax = sns.countplot(data = df, x='Gender')
plt.title("Gender Distribution")

for bars in ax.containers:
    ax.bar_label(bars)

# ax.bar_label(ax.conatiners[0])

plt.show()
```

# Conclusion: From the above charts we have analysed that the number of females is more than the number of males

```python
gb = df.groupby('ParentEduc').agg({'MathScore': 'mean', 'ReadingScore': 'mean', 'WritingScore': 'mean'})
```

```python
gb
```

```python
sns.heatmap(gb, annot = True)
plt.title("Relationship between Parent's Education and Student's Scores")
plt.show()
```

# Conclusion: From the above chart we have concluded that the eductaion of parents have good impact on the Student's Scores

```python
gb1 = df.groupby('ParentMaritalStatus').agg({'MathScore': 'mean', 'ReadingScore': 'mean', 'WritingScore': 'mean'})
```

```python
gb1
```

```python
sns.heatmap(gb1, annot = True)
plt.title("Relationship between Parent's Marital Status and Student's Scores")
plt.show()
```

# Conclusion: From the above chart we have seen that the ParentMaritalStatus have neglible impact on Student's Score

# Plotting A Pie Chart: Distribution by Ethnic Group

```python
mylist = [groupA['EthnicGroup'], groupB['EthnicGroup'], groupC['EthnicGroup'], groupD['EthnicGroup'], groupE['EthnicGroup']]
l = ['group A', 'group B', 'group C', 'group D', 'group E']
plt.figure(figsize=(8, 8))
plt.title("Distribution by Ethnic Group")
plt.legend(l, loc="upper left", bbox_to_anchor=(1, 0.5))
plt.pie(mylist, labels = l, autopct = '%1.2f%%')
plt.tight_layout()
plt.show()
```
